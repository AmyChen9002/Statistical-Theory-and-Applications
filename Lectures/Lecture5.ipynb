## For Loops

The above graph can also be obtained through for-loop as:

```python
ts_length = 100
epsilon_values = []   # empty list

for i in range(ts_length):
    e = np.random.randn()
    epsilon_values.append(e)

plt.plot(epsilon_values)
plt.show()
```

The format of for-loop:\
for $\textit{i}$ in range($n$):\
$\space$ $\space$ computation 

Another example:

```python
animals = ['dog', 'cat', 'bird']
for animal in animals:
    print("The plural of " + animal + " is " + animal + "s")
```

Let’s do one more application here.

In this application, we plot the balance of a bank account over time.

There are no withdraws over the time period, the last date of which is denoted by $T$.

The initial balance is $b_0$ and the interest rate is $r$.

The balance updates from period $t$ to $t+1$ according to $b_{t+1}=b_t(1+r)$.

In the code below, we generate and plot the sequence $b_0,b_1,...,b_T$.

Instead of using a Python list to store this sequence, we will use an array.

```python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty NumPy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```

## While-Loop

Another method of looping is using while-loop.\
So the example above can also be realized as the following:

```python
ts_length = 100
epsilon_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    epsilon_values.append(e)
    i = i + 1
plt.plot(epsilon_values)
plt.show()
```

The format of while loop:\
while $i<$ the range\
$\space$ $\space$ your code block

Keyword arguments are particularly useful when a function has a lot of arguments, in which case it’s hard to remember the right order.\

```python
def f(x, a=1, b=1):
    return a + b * x
```

The keyword argument values we supplied in the definition of $f$ become the default values

```python
f(2)
```

## Applications

Consider again this code from the beginning

```python
import numpy as np
import matplotlib.pyplot as plt
ts_length = 100
epsilon_values = []   # empty list

for i in range(ts_length):
    e = np.random.randn()
    epsilon_values.append(e)

plt.plot(epsilon_values)
plt.show()
```

We will break this program into two parts:

A user-defined function that generates a list of random variables.

The main part of the program that

calls this function to get data

plots the data

This is accomplished in the next program


```python
def generate_data(n):
    epsilon_values = []
    for i in range(n):
        e = np.random.randn()
        epsilon_values.append(e)
    return epsilon_values

data = generate_data(100)
plt.plot(data)
plt.show()
```

### Adding Conditions

Let’s make it slightly more useful by giving it the ability to return either standard normals or uniform random variables on $(0,1)$ as required.

```python
def generate_data(n, generator_type):
    ϵ_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        ϵ_values.append(e)
    return ϵ_values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```
